In Python, **lists** are useful for storing multiple items in a single variable. However, they donâ€™t associate values with meaningful keys. For example, suppose we are working with biomedical data. We might have one list containing the **names of patients** and another list containing their **systolic blood pressure values**, both ordered in the same sequence. 

If we want to retrieve a patient's blood pressure, we must loop through both lists simultaneously using the `zip()` function. This approach can quickly become inefficient and hard to manageâ€”especially when dealing with more than two attributes.

## ðŸ§¬ Why Use Dictionaries?

While handling patient data, we often need to manage **multiple attributes** such as:

- Name  
- Age  
- Gender  
- Body weight  
- Systolic blood pressure  
- Diagnosis  
- Prescribed drugs  
- Co-existing ailments  
- Allergies  

Managing all of these attributes using only lists becomes complex and error-prone. This is where **Python dictionaries** become incredibly useful.

Dictionaries allow us to store **key-value pairs**, where each key acts as a descriptive label for its associated value. This structure makes it easy to **access**, **modify**, **add**, or **delete** any attribute as the patientâ€™s treatment progresses.

---

## ðŸ§¾ General Characteristics of a Python Dictionary

- Dictionaries are defined using **curly brackets `{}`**.
- A dictionary may be **empty** or contain **one or more key-value pairs**.
- Each **key** represents an attribute (e.g., `"age"`, `"blood_pressure"`) and is followed by a **colon `:`** separating it from its corresponding **value**.
- Each key-value pair is separated from the next by a **comma `,`**.

---

# Working with Dictionaries

````python
# Let us practice this in code now. We write a simple dictionary below
drug_class = {
    "aspirin": "NSAID",
    "ciprofloxacin": "fluoroquinolone antibiotic",
    "amoxicillin": "penicillin antibiotic",
    "fexofenadine": "antihistamine"
}
````

## Accessing Key-Value Pairs

````python
# We may access and print the key value pairs by asking the key
print("The class of the drug mentioned as the index number one of the list named 'drug_class' is: ", drug_class["aspirin"])
print("\nThe class of the drug mentioned as the index number two of the list named 'drug_class' is: ", drug_class["ciprofloxacin"].title())
print("\nThe class of the drug mentioned as the index number three of the list named 'drug_class' is: ", drug_class["amoxicillin"].title())
print("\nThe class of the drug mentioned as the index number four of the list named 'drug_class' is: ", drug_class["fexofenadine"].title())
````

````python
# We may access both the key as well as its value using items() method
for key, value in drug_class.items():
    print(f"\nThe name of the drug is {key} and its class is {value}.")
````

````python
# To make this code even more better, we also add index number by combining the enumerate function with items function
for index, (key, value) in enumerate(drug_class.items(), start = 1):
    print(f"\nThe name of hte drug at position {index} is {key.upper()} and its class is {value.upper()}.")
````

### Get() Function to Access Key-Value Pair

You have seen in the above output that we may get the value of any key present within the dictionary by simply mentioning its key in square brackets next to the dictionary name. However, it is really a problem if we are not sure about the presence of the particular key-value pair within the dictionary. If the dictionary does not contain the key that we are inquiring, the code mentioned above to access the dictionary items will produce error. A better way is to use get() function to access any key. Let us practice it in the code below:

````python
print("The Dictionary Named 'drug_class' is: ", drug_class)
print(drug_class["indomethacin"])
````

Since there is no key named 'indomethacin' in this dictionary, we get an error. A better way to inquire such key is using get() function. What is may do is write dictionary name, followed by .get() instead of the name of the key within brackets. 
`dictionary_name.get()`
Then, we provide the name of the key as argument in the get() function. Additionally, we may provide a message as the second argument of get() function if the item is not present in the list. However, this second argument is optional. Even if we don't provide this second argument and the key is not present in the dictionary, we will get a None output instead of an error.

````python
print(drug_class.get('indomethacin', 'The drug is not listed as a key in this dictionary'))
````

````python
print(drug_class.get('ipecac'))
````

## Adding a Key to a Dictionary

````python
# We can simply add a new key-value pair to an existing dictionary by defining the new pair
print("The original dictionary is: ", drug_class)
drug_class["omeprazole"] = "proton pump inhibitor"
drug_class["diclofenac"] = "NSAID"
drug_class["acetaminophen"] = "antipyretic"
drug_class["insulin"] = "hormone"
print("\nThe updated dictionary is: ", drug_class)
````

Unlike Python lists, dictionaries are not indexed in the same way. Although modern Python versions (3.7 and above) preserve insertion order, dictionaries still do not support positional indexing. Therefore, we cannot use the `insert()` function to add a key-value pair at a specific index, as we can with lists.

In our programming journey, we may encounter situations where a program starts with an empty dictionary and gradually builds it by adding key-value pairs as the logic progresses. This dynamic construction of dictionaries is a common and powerful pattern in Python programming.


````python
drugs = {}
name_1 = input("Patient's Name: ")
drug_1 = input("drug prescribed? ")
name_2 = input("Patient's Name: ")
drug_2 = input("drug prescribed?")

drugs[name_1] = drug_1
drugs[name_2] = drug_2
````

````python
print(drugs)
````

````python
for key, value in drugs.items():
    print(f"\nMr. {key} was prescribed {value} by the duty doctor in OPD today.")
````

## Deleting a Key from a Dictionary

````python
# We may delete any key-value pair from the dictionay using the del function
print("\nThe updated dictionary is: ", drug_class)
del drug_class["insulin"]
print("\nThe final dictionary is: ", drug_class)
````

# Looping Through a Dictionary

### Looping with keys only

````python
# We can simply loop through a dictionary's keys using two different methods, a simple for loop or by using key() method in for loop. We get a similar output in both these methods
drugs = {
    "Paracetamol": "Antipyretic / Analgesic",
    "Amoxicillin": "Antibiotic (Penicillin class)",
    "Ibuprofen": "NSAID (Non-Steroidal Anti-Inflammatory Drug)",
    "Ciprofloxacin": "Antibiotic (Fluoroquinolone)",
    "Omeprazole": "Proton Pump Inhibitor (PPI)",
    "Metformin": "Antidiabetic (Biguanide)",
    "Lisinopril": "ACE Inhibitor (Antihypertensive)",
    "Simvastatin": "Statin (Lipid-lowering agent)",
    "Azithromycin": "Antibiotic (Macrolide)",
    "Losartan": "ARB (Angiotensin II Receptor Blocker)",
    "Cetirizine": "Antihistamine (H1 blocker)",
    "Prednisolone": "Corticosteroid (Anti-inflammatory)",
    "Amlodipine": "Calcium Channel Blocker (Antihypertensive)",
    "Pantoprazole": "Proton Pump Inhibitor (PPI)",
    "Atorvastatin": "Statin (Lipid-lowering agent)",
    "Salbutamol": "Beta-2 Agonist (Bronchodilator)",
    "Clopidogrel": "Antiplatelet agent",
    "Doxycycline": "Antibiotic (Tetracycline class)",
    "Levothyroxine": "Thyroid Hormone Replacement",
    "Albendazole": "Anthelmintic (Anti-parasitic)"
}

# Method 1, looping via a simple for loop
for drug in drugs:
    print(f"\nThe name of the drug is {drug}.")
````

````python
# Method 2, using a key() function in for loop to enhance more readability and clearity in our code
for drug in drugs.keys():
    print(f"\nThe name of the drug is {drug}.")
````

````python
# The presentation looks a bit wierd. So we may enhance its output as below
for index, value in enumerate(drugs.keys(), start = 1):
    print(f"\nThe Name of Drug in the Dictionary Named 'Drugs' at Place {index} is {value}.")
````

#### Looping through Keys with Sorted() function

````python
# Alternativley, we may loop a dictionary's keys in a order instead of their index number
for drug in sorted(drugs.keys()):
    print(f"\nThe drug in the dictionary named 'Drugs' is {drug}")
````

### Looping With Values Only

````python
# We may loop a dictionary with its values only
print("The different classes of drugs included in the dictionary named 'Drugs' include: ")
for value in drugs.values():
    print(f"\n{value}")
````

### Using set() function to see only unique values in a dictionary, skipping the repeated values

There could be some values in a dictionary that are repeated. If you need to see only the unique values, you may make a set of the dictionary values. You may get confused between a dictionary and a set, since both of these are written within curly brackets. However, the sets don't have key-value pairs and the items within sets are separated by commas just like we see in lists. The important characteristic of sets is that they include only unique values. The repeated values are automatically excluded from the sets.

`dictionary = {"key" : "value", "key_2" : "value_2", "key_3" : "value_3"}`

`set = {"value", "value_2", "value_3"}`

````python
print(drugs)
````

````python
print(set(drugs)) # This only prints the keys
print("\nThe total number of unique keys in the dictionary 'drugs' is: ", len(set(drugs)))
````

````python
print(set(drugs.values())) # This only prints the values
print("\nThe total number of unique values in the dictionary 'drugs' is: ", len(set(drugs.values())))
````

### Looping with both keys and values

````python
# If we wish to loop through both the keys and the values within a dictionary, we may use item() function
print("The Original Dictionary is: ", drugs)
print("\n DRUG | CLASS")
for key, value in drugs.items():
    print(f"\n{key} | {value}")
````
